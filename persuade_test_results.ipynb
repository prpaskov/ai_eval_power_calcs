```python
import pandas as pd 
from aggregate_scores import AggregateScores
from analyze_scores import AnalyzeScores
from run_power_calcs import *
from collections import Counter
```

```python
path = '/Users/patriciapaskov/Documents/GitHub/ai_eval_power_calcs/'
persuade =  AggregateScores(eval_name = "persuade",
                                    max_score = 6.0, ##if this is question-specific, will need to tweak the code
                                    comparison_judge_id = 'baseline',
                                    question_id_col = 'item_id',
                                    folder_path= 'data/decomp_judges',
                                    verbose = False,
                                    ground_truth_col = 'holistic_essay_score',
                                    judge_score_col = 'judge_score')
persuade_wide = persuade.load_reshaped_df()
judges = [c for c in persuade_wide.columns if 'judge_' in c and c!='judge_name']
persuade_wide.head()
```

```python
persuade_analyzer = AnalyzeScores( 
                 df = persuade_wide, 
                 test_cols = judges,
                comparison_col = 'baseline',
                draws_col = 'draws',
                cluster_col = None,
                bonferroni_correction = True,
                num_comparisons = 90,               ##update - how many test are being run against baseline?
                alpha = 0.05,
                beta = 0.20)

results_persuade = persuade_analyzer.run_multiple_tests()
persuade_analyzer.get_summary_table(results_persuade)
```

```python
persuade_latex_basic = persuade_analyzer.export_latex_table(results_persuade, 
                                         output_file="results/persuade_results_basic.tex",
                                         caption="Persuade Model Comparison Results")

latex_enhanced = persuade_analyzer.export_latex_table_enhanced(results_persuade,
                                        output_file="results/persuade_results_enhanced.tex", 
                                        caption="Persuade Comprehensive Model Comparison",
                                        include_ci=True,
                                        include_sample_size=True)
```
